```python
# Import required libraries

import pandas as pd

from pymongo import MongoClient
```

```python
# MongoDB Atlas connection

url = "mongodb+srv://mohit:mohit@cluster0.hvutxv4.mongodb.net/"

client = MongoClient(url)

print("MongoDB Atlas connection successful.")
```

```python
# Create database

db = client["Database"]

print("Database created successfully.")
```

```python
# Create collections

customers_collection = db["Customers"]

deliveries_collection = db["Deliveries"]

drivers_collection = db["Drivers"]

hubs_collection = db["Hubs"]

app_events_collection = db["App_events"]

print("Collections created successfully.")
```

```python
# Import datasets

customers = pd.read_csv(
    "https://raw.githubusercontent.com/mohitpreet9435-gif/Dataset/main/customers.csv"
)

deliveries = pd.read_csv(
    "https://raw.githubusercontent.com/mohitpreet9435-gif/Dataset/main/deliveries.csv"
)

drivers = pd.read_csv(
    "https://raw.githubusercontent.com/mohitpreet9435-gif/Dataset/main/drivers.csv"
)

hubs = pd.read_csv(
    "https://raw.githubusercontent.com/mohitpreet9435-gif/Dataset/main/hubs.csv"
)

app_events = pd.read_csv(
    "https://raw.githubusercontent.com/mohitpreet9435-gif/Dataset/main/app_events.csv"
)
```

```python
# Convert datasets to dictionaries

customers_data = customers.to_dict(
    orient="records"
)

deliveries_data = deliveries.to_dict(
    orient="records"
)

drivers_data = drivers.to_dict(
    orient="records"
)

hubs_data = hubs.to_dict(
    orient="records"
)

app_events_data = app_events.to_dict(
    orient="records"
)
```

```python
# Insert customer data

customers_collection.insert_many(
    customers_data
)

# Insert delivery data

deliveries_collection.insert_many(
    deliveries_data
)

# Insert driver data

drivers_collection.insert_many(
    drivers_data
)

# Insert hub data

hubs_collection.insert_many(
    hubs_data
)

# Insert app event data

app_events_collection.insert_many(
    app_events_data
)

print("Data inserted successfully.")
```

CRUD OPERATION 1 — CREATE

```python
# Insert new customer document

new_customer = {
    "customer_id": "C9999",
    "customer_type": "Premium",
    "home_zone": "Central",
    "loyalty_score": 88,
    "app_engagement_score": 91,
    "preferred_channel": "Mobile App",
    "account_status": "Active"
}

customers_collection.insert_one(
    new_customer
)

print("New customer inserted successfully.")
```

CRUD OPERATION 2 — READ

```python
# Read customer documents

results = customers_collection.find().limit(5)

for document in results:
    print(document)
```

CRUD OPERATION 3 — UPDATE

```python
# Update customer account status

customers_collection.update_one(
    {"customer_id": "C9999"},
    {
        "$set": {
            "account_status": "Inactive"
        }
    }
)

print("Customer document updated successfully.")
```

CRUD OPERATION 4 — DELETE

```python
# Delete customer document

customers_collection.delete_one(
    {"customer_id": "C9999"}
)

print("Customer document deleted successfully.")
```

MongoDB Query 1 — Delayed Deliveries

```python
# Find delayed deliveries

delayed_deliveries = deliveries_collection.find(
    {
        "delivery_status": "Delayed"
    }
)

for document in delayed_deliveries.limit(5):
    print(document)
```

Query 2 — High Severity Complaints

```python


complaints_collection = db["complaints"]


complaints = pd.read_csv(
    "https://raw.githubusercontent.com/mohitpreet9435-gif/Dataset/main/complaints.csv"
)


complaints_data = complaints.to_dict(
    orient="records"
)


complaints_collection.insert_many(
    complaints_data
)


high_severity = complaints_collection.find(
    {
        "severity": "High"
    }
)

for document in high_severity.limit(5):
    print(document)
```

Query 3 — Drivers with Low Ratings

```python
# Find low-rated drivers

low_rated_drivers = drivers_collection.find(
    {
        "driver_rating": {
            "$lt": 3
        }
    }
)

for document in low_rated_drivers.limit(5):
    print(document)
```

Query 4 — App Events with High API Latency

```python
# High latency app events

high_latency = app_events_collection.find(
    {
        "api_latency_ms": {
            "$gt": 500
        }
    }
)

for document in high_latency.limit(5):
    print(document)
```

Aggregation — Delivery Status Analysis

```python
# Aggregation pipeline

pipeline = [
    {
        "$group": {
            "_id": "$delivery_status",
            "total_deliveries": {
                "$sum": 1
            }
        }
    }
]

results = deliveries_collection.aggregate(
    pipeline
)

for document in results:
    print(document)
```

MongoDB Aggregation — Average Fuel Cost by Hub

```python
# Average fuel cost by hub

pipeline = [
    {
        "$group": {
            "_id": "$hub_id",
            "average_cost": {
                "$avg": "$fuel_or_charge_cost"
            }
        }
    }
]

results = deliveries_collection.aggregate(
    pipeline
)

for document in results:
    print(document)
```

```python
# Create indexes

deliveries_collection.create_index(
    "delivery_status"
)

deliveries_collection.create_index(
    "hub_id"
)

customers_collection.create_index(
    "customer_id"
)

print("Indexes created successfully.")
```
